Analise de dados sobre covid nos Estados Brasileiros

```python
#importando a biblioteca Pandas

import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('https://raw.githubusercontent.com/wcota/covid19br/master/cases-brazil-states.csv')
```

```python
#visualizando as 5 primeiras colunas
df.head()
```

```python
#selecionando colunas do dataframe
df = df.loc[:,["date","country","city","state","tests","totalCases","newCases","vaccinated","deaths"]]
df.head()
```

```python
#Renomeando colunas do dataframe
df = df.rename(columns={"date":"Data","country":"Pais", "city":"Cidade","state":"Estado","tests": "Testes","totalCases":"Total de casos","newCases":"Novos casos","vaccinated":"Vacinados","deaths":"Mortos" })
df.head()
```

```python
#informa√ß√µes estatistica
df.describe()
```

```python
#convertendo a nota√ß√£o cientifica
pd.options.display.float_format ='{:20,.2f}'.format
df.head()
```

```python
#Estados confirmados
df.Estado.unique()
```

```python
#Estados e total de casos
```

```python
#total de casos por estados
df.groupby("Estado")["Total de casos"].sum().sort_values(ascending = False)
```

```python
#Grafico do total de casos nos Estados
df.groupby("Estado")["Total de casos"].sum().sort_values(ascending = True).plot.bar(title ="Total Casos por Estados")
plt.xlabel ("Estado")
plt.ylabel("Total de casos");
plt.grid()
```
